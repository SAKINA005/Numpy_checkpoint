To create a program that calculates the grades of students using Numpy arrays.

Description: 

In this exercise, you will create a program that calculates the grades of students using Numpy arrays. The program will ask the user to enter the number of students, the number of subjects, and the marks of each student in each subject. It will then calculate the total marks, percentage, and grade for each student using Numpy arrays. The program will display the result for each student in a tabular format.


Instructions

Instructions:

Import the Numpy library.
Ask the user to enter the number of students and the number of subjects.
Create a Numpy array to store the marks of each student in each subject. The array should have the shape (number of students, number of subjects).
Ask the user to enter the marks of each student in each subject.
Calculate the total marks for each student using the sum() function of Numpy.
Calculate the percentage for each student using the total marks and the formula: percentage = (total marks / (number of subjects * 100)) * 100
Calculate the grade for each student using the percentage and the following grading system:
90% or above: A+
80% or above: A
70% or above: B+
60% or above: B
50% or above: C
below 50%: F
Display the result for each student in a tabular format with columns for student name, total marks, percentage, and grade.

```python
import numpy as np
def grade_student(percentage):
    if percentage >=90:
        return "THE GRADE IS A+"
    elif percentage >=80:
        return "THE GRADE IS A"
    elif percentage >=70:
        return "THE GRADE IS B+"
    elif  percentage >=60:
        return "THE GRADE IS B"
    elif  percentage >=50:
        return "THE GRADE IS C"   
    elif percentage < 50:
        return "THE GRADE IS F"
    


while True:
    try:
        nb_students=int(input("VEUILLEZ ENTRER LE NOMBRE D'ELEVES"))
        nb_subjects=int(input("VEUILLEZ ENTRER LE NOMBRE DE SUJETS"))
        break
    except ValueError:
        print("VEUILLEZ ENTRER DES ENTIERS")
store_marks=np.full((nb_students,nb_subjects),None)

students_name= []
for i in range(nb_students):
    name=input(f"Entrez le nom de l'élève{i+1}")
    students_name.append(name)

subjects_name= []
for j in range(nb_subjects):
    subject=input(f"Entrez le nom du sujet{j+1}")
    subjects_name.append(subject)

for i in range(nb_students):
    for j in range(nb_subjects) :
        mark=int(input(f"VEUILLEZ ENTREZ LA NOTE DE L'ELEVE {students_name[i]} pour le sujet {subjects_name[j]}"))
        while True:
            try:
                if 0 <= mark <=100:
                    store_marks[i,j]=mark
                    break
                else:
                    print("VEUILLEZ ENTREZ UNE NOTE COMPRISE ENTRE O ET 100")       
            except ValueError:
                print("VEUILLEZ ENTRER UN NOMBRE VALIDE")
                
total_marks=np.sum(store_marks,axis=1)
percentage=(total_marks / (nb_subjects * 100)) * 100 
grade=[grade_student(p) for p in percentage]
print("\nRésultats :")
print("{:<12} {:<12} {:<12} {:<12}".format("Élève", "Total", "Pourcentage", "Note"))
for i in range(nb_students):
        print("{:<12} {:<12} {:<12.2f} {:<12}".format(students_name[i], total_marks[i], percentage[i], grade[i]))





```
